```python
import os
import shutil
import pandas as pd

csv_file = "/kaggle/input/pad-uefs-20/PAD-UFES-20_metadata.csv"
image_folder = "/kaggle/input/part1datasett/kaggle/working/PAD-UFES-20_images_part_11forzip"
destination_folder = '/kaggle/working/Dataset'

df = pd.read_csv("/kaggle/input/pad-uefs-20/PAD-UFES-20_metadata.csv")


for index,row in df.iterrows():
    image_id = str(row['img_id'])
    diagnosis = str(row['diagnostic'])
    
    destination_folder = os.path.join(destination_folder,diagnosis)
    
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)
    
    source_path = os.path.join(image_folder,image_id)
    destination_path = os.path.join(destination_folder,image_id)
    
    shutil.copy(source_path,destination_path)
    
    i = 0
    
    print(f"image {i} is saved on {destination_folder}")
    
```
